# Task 1

### What makes an AI agent different from a traditional machine learning model?

An AI agent is different from a traditional machine learning model because it can make decisions, plan actions, and complete tasks on its own instead of only making predictions. A traditional machine learning model usually gives an output based on the input it receives but cannot continue working without user instructions.

**Example:** In Instagram, a traditional ML model recommends reels based on your interests. An AI agent could also organize your messages, suggest replies, and remind you about important conversations without needing you to ask each time.

# Task 2

### Three characteristics of agent-based AI systems

#### 1. Autonomy
An AI agent can perform tasks independently without requiring constant user input.

**Example:** In Swiggy, the agent can automatically reorder your favorite meal every Friday evening if it notices your regular ordering pattern.

#### 2. Decision Making
An AI agent can analyze different options and choose the best one based on available information.

**Example:** In Flipkart, the agent can compare prices, ratings, delivery time, and discounts to recommend the best product.

#### 3. Goal-Oriented Behavior
An AI agent works towards completing a specific goal by performing multiple related actions.

**Example:** In BookMyShow, the agent can search for a movie, choose the best seats, apply coupons, and remind the user before the show starts.

# Task 3

### How AutoGPT or ReAct Agents work

AutoGPT and ReAct Agents are AI systems that can think through a problem, divide it into smaller steps, collect information when needed, make decisions, and perform actions with minimal human guidance. Instead of waiting for instructions after every step, they continue working until the assigned goal is completed.

### Use Case (Spotify)

A Spotify AI agent could automatically create a playlist based on the user's listening history, remove songs that are frequently skipped, add new songs from favorite artists, and update the playlist every week without the user requesting it.
