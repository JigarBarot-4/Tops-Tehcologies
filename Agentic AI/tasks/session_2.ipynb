# Task 1

### Create a function `perceive_input()` that takes a user's message and returns the detected intent and keywords.

```python
def perceive_input(message):
    message = message.lower()

    intent = "unknown"
    keywords = []

    if "show" in message:
        intent = "show"
    elif "play" in message:
        intent = "play"
    elif "book" in message:
        intent = "book"

    if "spotify" in message:
        keywords.append("spotify")

    if "youtube" in message:
        keywords.append("youtube")

    if "bookmyshow" in message:
        keywords.append("bookmyshow")

    return {
        "intent": intent,
        "keywords": keywords
    }

# Example
user = "Show me trending songs on Spotify"

result = perceive_input(user)

print(result)
```

# Task 2

### Create a function `decide_next_action()` that decides what action should be performed.

```python
def decide_next_action(data):

    if data["intent"] == "show" and "spotify" in data["keywords"]:
        return "fetch_spotify_trending"

    elif data["intent"] == "book" and "bookmyshow" in data["keywords"]:
        return "search_movie_bookmyshow"

    elif data["intent"] == "play" and "youtube" in data["keywords"]:
        return "play_youtube_video"

    else:
        return "No action found"

# Example
action = decide_next_action(result)

print(action)
```

# Task 3

### Create an `execute_action()` function and complete the Perceive → Think → Act loop.

```python
def execute_action(action):

    if action == "fetch_spotify_trending":
        print("Fetching trending songs from Spotify...")

    elif action == "search_movie_bookmyshow":
        print("Searching movies on BookMyShow...")

    elif action == "play_youtube_video":
        print("Playing video from YouTube...")

    else:
        print("Action not available.")


# Complete Agent Loop

user = input("Enter your request: ")

data = perceive_input(user)

print("Perceived Data:", data)

action = decide_next_action(data)

print("Next Action:", action)

execute_action(action)
```

### Mapping the Perceive → Think → Act loop to an LLM-based Agent

In an LLM-based agent, the **perceive_input()** function would understand the user's message using natural language instead of checking fixed keywords. The **decide_next_action()** function would use the LLM to choose the best action based on the user's request. The **execute_action()** function would connect with real apps or APIs, and the LLM would generate a clear and natural response for the user.
