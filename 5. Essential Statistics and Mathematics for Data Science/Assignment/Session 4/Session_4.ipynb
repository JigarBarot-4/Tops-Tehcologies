# Question 1: Given the daily order counts for a Swiggy delivery partner over 10 days: [12, 15, 17, 14, 13, 16, 200, 18, 14, 15], plot a simple graph (hand-drawn or using any tool) and describe whether the data is left skewed, right skewed, or symmetrical. Explain your reasoning based on the shape.

## Answer

```python
import matplotlib.pyplot as plt

orders = [12, 15, 17, 14, 13, 16, 200, 18, 14, 15]

plt.plot(orders, marker='o')
plt.title("Daily Swiggy Orders")
plt.xlabel("Day")
plt.ylabel("Number of Orders")
plt.grid(True)
plt.show()
```

The data is **right skewed** because there is one very large value (200) compared to the other values. Most of the order counts are between 12 and 18, while the value 200 creates a long tail on the right side of the graph.

# Question 2: Take the following list of Instagram influencer follower counts (in thousands): [5, 7, 8, 8, 9, 10, 12, 15, 95]. Calculate the mean and median, then state which is higher and what that tells you about the skewness of the data.

## Answer

```python
import statistics

followers = [5, 7, 8, 8, 9, 10, 12, 15, 95]

mean = statistics.mean(followers)
median = statistics.median(followers)

print("Mean:", round(mean, 2))
print("Median:", median)
```

# Question 4: Given this dataset of Flipkart product ratings: [3, 3, 4, 4, 4, 5, 5, 5, 5, 5], draw a rough graph and state if the distribution is symmetrical, left skewed, or right skewed. Justify your answer in one line.

## Answer

```python
import matplotlib.pyplot as plt

ratings = [3, 3, 4, 4, 4, 5, 5, 5, 5, 5]

plt.hist(ratings, bins=3)
plt.title("Flipkart Product Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

The distribution is **left skewed** because most ratings are 5, while only a few ratings are 3, creating a tail towards the left.
