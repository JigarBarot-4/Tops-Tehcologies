# Question 1: Write a Python function bayes_posterior(prior, likelihood, evidence) that calculates and returns the posterior probability using Bayes theorem formula for given values.

## Answer

```python
def bayes_posterior(prior, likelihood, evidence):
    posterior = (likelihood * prior) / evidence
    return posterior

prior = 0.2
likelihood = 0.8
evidence = 0.5

result = bayes_posterior(prior, likelihood, evidence)

print("Posterior Probability:", result)
```

# Question 2: Given the following scenario: On Swiggy, 10% of restaurants are 'premium', and 80% of premium restaurants offer a discount, while only 20% of regular restaurants do. If you see a discount, use Bayes theorem in code to find the probability that the restaurant is premium.

## Answer

```python
# Prior probability of a restaurant being premium
prior = 0.10

# Probability of discount given premium restaurant
likelihood = 0.80

# Probability of a regular restaurant
regular = 0.90

# Probability of discount given regular restaurant
regular_discount = 0.20

# Evidence: Probability of getting a discount
evidence = (likelihood * prior) + (regular_discount * regular)

# Bayes Theorem
posterior = (likelihood * prior) / evidence

print("Probability that the restaurant is premium:", round(posterior, 3))
```

# Question 3: Simulate a basic spam detection filter: Given a list of 20 email subjects (some spam, some not), and the probability that 'free' appears in spam vs non-spam emails, write code to calculate the probability an email is spam if it contains 'free', using Bayes theorem.

## Answer

```python
emails = [
    "Free Offer", "Meeting Today", "Win Money", "Free Gift",
    "Project Update", "Free Recharge", "Invoice", "Discount Offer",
    "Team Meeting", "Free Coupons", "Salary Update", "Limited Offer",
    "Free Trial", "Holiday Notice", "Congratulations",
    "Free Movie", "Weekly Report", "Shopping Deal",
    "Free Bonus", "Client Feedback"
]

# Bayes values
prior_spam = 0.4
likelihood = 0.8          # P(Free | Spam)
evidence = 0.5            # P(Free)

posterior = (likelihood * prior_spam) / evidence

print("Probability that an email is spam if it contains 'Free':", round(posterior, 2))
```
