# Question 2: Given this sample data from a music app: ['Pop', 'Rock', 'Jazz', 'Hip-Hop', 'Pop', 'Rock', 'Jazz', 'Pop'], classify the type of data and explain your reasoning in 2-3 lines.

## Answer

The given data is **categorical (nominal) data** because it contains music genres instead of numbers. These values represent different categories and cannot be measured or arranged in any numerical order. The data is useful for analyzing users' music preferences.
