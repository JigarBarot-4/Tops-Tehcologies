# Question 1: Given two 2x2 matrices representing the number of likes and comments on two Instagram posts over two days, write Python code to multiply them and display the resulting matrix.

## Answer

```python
import numpy as np

likes = np.array([
    [100, 120],
    [150, 180]
])

comments = np.array([
    [10, 15],
    [20, 25]
])

result = np.matmul(likes, comments)

print("Result of Matrix Multiplication:")
print(result)
```

# Question 2: Create a simple Python function linear_regression_predict(x, m, c) that takes an input value x, slope m, and intercept c, and returns the predicted output y for a linear regression equation y = mx + c.

## Answer

```python
def linear_regression_predict(x, m, c):
    y = m * x + c
    return y

print(linear_regression_predict(5, 4, 10))
```

# Question 3: Imagine you are predicting the delivery time for Zomato orders based on distance. Given three distances (2km, 5km, 8km) and a linear regression equation y = 4x + 10 (where y is minutes), use your function from the previous task to calculate the predicted delivery time for each distance.

## Answer

```python
def linear_regression_predict(x, m, c):
    return m * x + c

distances = [2, 5, 8]

for distance in distances:
    time = linear_regression_predict(distance, 4, 10)
    print(f"Distance: {distance} km -> Delivery Time: {time} minutes")
```

# Question 4: Explain in your own words what the slope and intercept represent in the context of Flipkart's product price prediction (e.g., predicting price based on product rating). Write 2-3 sentences for each term.

## Answer

**Slope:**

The slope shows how much the predicted product price changes when the product rating increases by one unit. A positive slope means higher-rated products are expected to have higher prices.

**Intercept:**

The intercept is the starting price predicted when the product rating is zero. It is the base value of the prediction before considering the effect of the rating.
