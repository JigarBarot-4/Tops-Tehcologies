# Question 1: Create a Python script called playlist_stats.py that takes a list of daily Spotify song play counts (e.g., [120, 135, 150, 200, 120, 90, 200]) and calculates the mean (average) number of plays.

## Answer

```python
# playlist_stats.py

plays = [120, 135, 150, 200, 120, 90, 200]

mean = sum(plays) / len(plays)

print("Average number of plays:", mean)
```

# Question 2: Given the following array of delivery times (in minutes) for Zomato orders: [30, 25, 40, 35, 30, 45, 30], write code to find the median delivery time and print it.

## Answer

```python
import statistics

delivery_times = [30, 25, 40, 35, 30, 45, 30]

median_time = statistics.median(delivery_times)

print("Median delivery time:", median_time)
```

# Question 3: Write a function most_common_rating(ratings) that takes a list of Flipkart product ratings (e.g., [5, 4, 4, 3, 5, 4, 2, 4]) and returns the mode (most frequent rating).

## Answer

```python
import statistics

def most_common_rating(ratings):
    return statistics.mode(ratings)

ratings = [5, 4, 4, 3, 5, 4, 2, 4]

print("Most common rating:", most_common_rating(ratings))
```

# Question 4: Given three lists representing YouTube video views for three different channels, compare the mean, median, and mode for each channel, and decide which channel's data is most affected by outliers.

## Answer

```python
import statistics

channel_1 = [100, 120, 110, 105, 115]
channel_2 = [200, 210, 220, 200, 230]
channel_3 = [100, 120, 110, 105, 5000]

channels = {
    "Channel 1": channel_1,
    "Channel 2": channel_2,
    "Channel 3": channel_3
}

for name, views in channels.items():
    print(name)
    print("Mean:", statistics.mean(views))
    print("Median:", statistics.median(views))
    print("Mode:", statistics.mode(views))
    print()
```
