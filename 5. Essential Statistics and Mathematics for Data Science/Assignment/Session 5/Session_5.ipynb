# Question 2: Write a Python script that simulates random sampling: given a list of 1000 fake user IDs (e.g., user1 to user1000), select 50 random users to simulate a survey sample. Print the selected user IDs.

## Answer

```python
import random

# Create 1000 fake user IDs
users = [f"user{i}" for i in range(1, 1001)]

# Select 50 random users
sample_users = random.sample(users, 50)

print("Selected Users:")
print(sample_users)
```

# Question 3: Imagine you are analyzing food delivery times on Swiggy. Explain in 2-3 sentences why you would use sampling instead of collecting data from every single order ever made.

## Answer

Sampling saves both time and cost because collecting data from every order is very difficult. A properly selected sample can still give a good idea of the overall delivery performance and help make reliable decisions.

# Question 4: Given this scenario: A survey about music preferences is only sent to users who recently streamed Bollywood songs on Spotify. Identify and explain the type of bias this introduces in the sample.

## Answer

This introduces **selection bias** because only users who listen to Bollywood songs are included in the survey. Users who prefer other music genres are ignored, so the results will not represent all Spotify users.

# Question 5: Choose any two types of sampling (for example: random sampling, stratified sampling, systematic sampling, cluster sampling). For each, describe how you would use it to select a sample of users from BookMyShow to study movie booking habits.

## Answer

**1. Random Sampling**

Randomly select users from the complete list of BookMyShow users. Every user has an equal chance of being selected for the survey.

**2. Stratified Sampling**

Divide users into groups such as students, working professionals, and senior citizens. Then randomly select users from each group so that every category is represented in the sample.
