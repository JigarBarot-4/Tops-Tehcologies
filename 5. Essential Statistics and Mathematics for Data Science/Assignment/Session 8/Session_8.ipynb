# Question 1: Write a Python function calculate_probability(event_count, total_outcomes) that returns the probability of an event, and use it to find the probability of getting 'Heads' when tossing a coin.

## Answer

```python
def calculate_probability(event_count, total_outcomes):
    return event_count / total_outcomes

# Probability of getting Heads
probability = calculate_probability(1, 2)

print("Probability of getting Heads:", probability)
```

# Question 2: Simulate rolling a six-sided dice 100 times using Python's random module, count how many times a '6' appears, and calculate the experimental probability of getting a '6'.

## Answer

```python
import random

count_six = 0

for i in range(100):
    roll = random.randint(1, 6)
    if roll == 6:
        count_six += 1

probability = count_six / 100

print("Number of times 6 appeared:", count_six)
print("Experimental Probability:", probability)
```

# Question 3: Given an array of genres from your last 20 Spotify songs (like ['pop', 'rock', 'pop', 'hiphop', ...]), write code to calculate the probability of randomly picking a 'pop' song from your recent history.

## Answer

```python
songs = [
    "pop", "rock", "pop", "hiphop", "jazz",
    "pop", "rock", "pop", "classical", "pop",
    "hiphop", "rock", "pop", "jazz", "pop",
    "rock", "pop", "pop", "hiphop", "pop"
]

pop_count = songs.count("pop")
probability = pop_count / len(songs)

print("Probability of selecting a pop song:", probability)
```

# Question 4: Write a function that, given two arrays — one for users who ordered food on Zomato and one for users who ordered dessert — calculates the probability that a randomly selected user ordered both food and dessert (i.e., intersection over total unique users).

## Answer

```python
def probability_both(food_users, dessert_users):
    food_set = set(food_users)
    dessert_set = set(dessert_users)

    both = food_set.intersection(dessert_set)
    total_users = food_set.union(dessert_set)

    return len(both) / len(total_users)

food = ["U1", "U2", "U3", "U4", "U5"]
dessert = ["U3", "U4", "U5", "U6"]

print("Probability:", probability_both(food, dessert))
```

# Question 5: Explain with a short Python code example how conditional probability could be used to improve recommendations in an app like Flipkart (e.g., probability of buying headphones given user bought a phone).

## Answer

```python
# Users who bought a phone
phone_buyers = 100

# Users who bought both phone and headphones
both_buyers = 40

# Conditional Probability
probability = both_buyers / phone_buyers

print("P(Headphones | Phone) =", probability)
```

This means that 40% of users who bought a phone also bought headphones. Flipkart can use this information to recommend headphones to customers who purchase a phone.
