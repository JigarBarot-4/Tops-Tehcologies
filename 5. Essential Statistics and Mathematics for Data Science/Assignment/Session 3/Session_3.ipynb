# Question 2: Write a Python function called calculate_standard_deviation(scores) that takes a list of exam scores and returns the standard deviation rounded to two decimal places.

## Answer

```python
import statistics

def calculate_standard_deviation(scores):
    return round(statistics.pstdev(scores), 2)

scores = [75, 80, 85, 90, 95]

print("Standard Deviation:", calculate_standard_deviation(scores))
```

# Question 3: Imagine two friends track their daily spending on Zomato over 5 days. Friend A spends [200, 200, 200, 200, 200] and Friend B spends [100, 300, 150, 400, 50]. Calculate the standard deviation for both and explain in 2-3 lines who has more consistent spending and why.

## Answer

```python
import statistics

friend_a = [200, 200, 200, 200, 200]
friend_b = [100, 300, 150, 400, 50]

std_a = statistics.pstdev(friend_a)
std_b = statistics.pstdev(friend_b)

print("Friend A Standard Deviation:", round(std_a, 2))
print("Friend B Standard Deviation:", round(std_b, 2))
```

Friend A has more consistent spending because the amount spent is the same every day, giving a standard deviation of 0. Friend B's spending changes a lot from day to day, so the standard deviation is much higher.

# Question 4: You are given the monthly salary data of 6 employees in a startup: [28000, 29000, 27000, 60000, 26500, 27500]. Calculate the variance and standard deviation, then interpret in 2-3 lines what the high or low spread means for this team.

## Answer

```python
import statistics

salaries = [28000, 29000, 27000, 60000, 26500, 27500]

variance = statistics.pvariance(salaries)
std_dev = statistics.pstdev(salaries)

print("Variance:", round(variance, 2))
print("Standard Deviation:", round(std_dev, 2))
```

The standard deviation is high because one employee earns much more than the others. This means there is a large spread in the salary data, and the salaries are not evenly distributed across the team.
