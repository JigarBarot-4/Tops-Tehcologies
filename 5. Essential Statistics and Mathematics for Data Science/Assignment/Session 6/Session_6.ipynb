# Question 1: Write down the null hypothesis (H0) and alternative hypothesis (H1) for the following scenario: You believe that people spend more time on Instagram than on YouTube daily.

## Answer

**Null Hypothesis (H₀):** People spend the same amount of time or less time on Instagram than on YouTube daily.

**Alternative Hypothesis (H₁):** People spend more time on Instagram than on YouTube daily.

# Question 2: Given the data: Group A (users shown a new Zomato home page) had an average order value of ₹350 from 30 users, and Group B (old home page) had an average order value of ₹320 from 30 users. Assume both groups have a standard deviation of ₹50. Calculate the p-value using a two-sample t-test (you may use Python, R, or an online calculator), and decide whether to accept or reject the null hypothesis at a 0.05 significance level.

## Answer

```python
from scipy.stats import ttest_ind_from_stats

mean1 = 350
std1 = 50
n1 = 30

mean2 = 320
std2 = 50
n2 = 30

t_stat, p_value = ttest_ind_from_stats(
    mean1=mean1, std1=std1, nobs1=n1,
    mean2=mean2, std2=std2, nobs2=n2,
    equal_var=True
)

print("T-statistic:", round(t_stat, 3))
print("P-value:", round(p_value, 4))
```

Since the p-value (0.0236) is less than 0.05, we reject the null hypothesis. This means the new Zomato home page has a statistically significant effect on the average order value.
