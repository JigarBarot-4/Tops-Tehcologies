# Question 1: Create a Python list to represent a playlist of 5 song durations in seconds (e.g., [210, 180, 245, 200, 175]). Print the total duration of the playlist.

## Answer

```python
playlist = [210, 180, 245, 200, 175]

total_duration = sum(playlist)

print("Total Playlist Duration:", total_duration, "seconds")
```

# Question 2: Represent a 3x3 matrix in Python to store the ratings (1-5) given by 3 friends to 3 different food items on Swiggy. Print the matrix and display the rating each friend gave to the second food item.

## Answer

```python
ratings = [
    [5, 4, 3],
    [4, 5, 5],
    [3, 4, 5]
]

print("Ratings Matrix:")
for row in ratings:
    print(row)

print("\nRatings for the second food item:")
for i in range(len(ratings)):
    print("Friend", i + 1, ":", ratings[i][1])
```

# Question 3: Given a matrix representing the number of orders for 4 different dishes across 3 days (rows: days, columns: dishes), write code to calculate and print the total orders for each dish.

## Answer

```python
orders = [
    [10, 15, 20, 12],
    [14, 18, 16, 10],
    [12, 20, 18, 15]
]

for i in range(4):
    total = 0
    for row in orders:
        total += row[i]
    print("Total orders for Dish", i + 1, ":", total)
```

# Question 4: Imagine you have a table of 5 Flipkart products with columns for price and quantity sold. Represent this data as a matrix in Python, then write a function to calculate the total revenue for each product.

## Answer

```python
products = [
    [15000, 5],
    [800, 20],
    [1200, 15],
    [500, 30],
    [25000, 2]
]

def calculate_revenue(data):
    for i, product in enumerate(data):
        revenue = product[0] * product[1]
        print("Product", i + 1, "Revenue:", revenue)

calculate_revenue(products)
```
