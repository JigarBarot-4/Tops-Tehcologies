# Question 1: Create two Python lists: one representing the number of hours you spend on Instagram each day for a week, and another for the number of posts you make each day. Plot these as a scatter plot using matplotlib and visually describe whether the relationship looks like a positive, negative, or zero correlation.

## Answer

```python
import matplotlib.pyplot as plt

# Hours spent on Instagram each day
hours = [1, 2, 2.5, 3, 4, 5, 6]

# Number of posts made each day
posts = [0, 1, 1, 2, 3, 4, 5]

plt.scatter(hours, posts)
plt.title("Instagram Usage vs Posts")
plt.xlabel("Hours Spent")
plt.ylabel("Number of Posts")
plt.grid(True)

plt.show()
```

The scatter plot shows a **positive correlation** because as the number of hours spent on Instagram increases, the number of posts also increases.

# Question 2: Given the following data for 6 friends: their monthly Zomato order count and their monthly Swiggy order count, calculate the correlation coefficient between the two lists using numpy's corrcoef(). Print the result and explain if the correlation is positive, negative, or zero.

## Answer

```python
import numpy as np

zomato = [8, 10, 12, 15, 18, 20]
swiggy = [6, 9, 11, 14, 16, 19]

correlation = np.corrcoef(zomato, swiggy)[0, 1]

print("Correlation Coefficient:", round(correlation, 2))
```

The correlation is **positive** because the value is close to 1. This means that when Zomato orders increase, Swiggy orders also tend to increase.
