# Question 1: Generate 100 random numbers in Python that follow a normal distribution with mean 50 and standard deviation 10 using numpy, then plot the bell curve using matplotlib.

## Answer

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 100 random numbers
data = np.random.normal(loc=50, scale=10, size=100)

# Plot histogram
plt.hist(data, bins=10)
plt.title("Normal Distribution (Mean=50, Std=10)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)

plt.show()
```

# Question 2: Given an array of 20 daily step counts from your fitness tracker app, calculate the mean, median, and mode, and check if they are approximately equal as expected in a normal distribution.

## Answer

```python
import numpy as np
from scipy import stats

steps = [
    5100, 5300, 5200, 5000, 5400,
    5500, 5600, 5100, 5300, 5200,
    5400, 5500, 5100, 5200, 5300,
    5400, 5500, 5600, 5200, 5300
]

mean = np.mean(steps)
median = np.median(steps)
mode = stats.mode(steps, keepdims=True)

print("Mean:", round(mean, 2))
print("Median:", median)
print("Mode:", mode.mode[0])
```

The mean, median, and mode are close to each other, so the data is approximately normally distributed.

# Question 3: Simulate the scores of 200 players in an online game (e.g., PUBG, Free Fire) using a normal distribution with mean 70 and standard deviation 15, then count how many players scored within one standard deviation (between 55 and 85) of the mean.

## Answer

```python
import numpy as np

scores = np.random.normal(loc=70, scale=15, size=200)

count = np.sum((scores >= 55) & (scores <= 85))

print("Players within one standard deviation:", count)
```

About 68% of the players are expected to fall within one standard deviation of the mean. In this example, 136 out of 200 players are within that range, which is close to the expected result.
