```python
class ATM :
    def __init__(self):
        self.name = None
        self.mobile_no = None
        self.pin = None
        self.balance = 0
        self.transaction_history = []
    
    def activate_card(self):
        self.name = input("Enter Your name :")
        self.mobile_no = input("Enter Mobile Number :")
        while True:
            if len(self.mobile_no) == 10 and self.mobile_no.isdigit() :
                print("Your card has been activated")
                break
            else:
                print("Enter Valid Mobile Number")
                self.mobile_no = input("Enter Mobile Number :")

    def pin_creation(self):
        while True:
            self.pin = input("Enter 4-Digit PIN :")
            pin_copy = input("Re-Enter your PIN :")
            if len(self.pin) == 4 and self.pin.isdigit() :
                if self.pin == pin_copy:
                    print("Your PIN is successfullly created !!")
                    break
                else:
                    print("PINs do not match")
            else:
                print("Instruction does not satisfied!")

    def initial_deposit(self):
        while True:
            amount = input("Enter minimum 1000 to active the card : ")
            if amount.isdigit():
                amount = int(amount)
                if amount >= 1000 :
                    self.balance += amount
                    print(f"Your current balance is {self.balance}")
                    self.transaction_history.append(f"{amount} deposited (initial deposit)")
                    break
                else:
                    print("Enter Amount 1000 or Above it ")
            else:
                print("Enter only digits ")

    def main_menu(self):
        while True:
            print("""
            1.Change PIN
            2.Deposit
            3.Withdraw
            4.Check Balance
            5.View Last 10 Transactions
            6.Exit                
        """)
            choice = input("Enter what you want to do in 1-6 format : ")
            
            if choice.isdigit():
                choice = int(choice)
                
                if choice == 1 :
                    current_pin = input("Enter your current PIN : ")
                    if current_pin == self.pin :
                        new_pin = input("Enter 4-digit new PIN :")
                        if len(new_pin) == 4 and new_pin.isdigit() :
                            conf_new_pin = input("Re-Enter your 4-digit PIN ; ")
                            if conf_new_pin == new_pin :
                                print("Your PIN has been successfully changed!!")
                                self.pin = conf_new_pin
                            else:
                                print("PINs Do not match!")
                        else:
                            print("Enter 4-digit PIN!!")
                    else:
                        print("Invalid PIN")

                elif choice == 2 :

                    ver_pin = input("Enter your PIN :")
                    if ver_pin == self.pin :
                        new_amount = input("Enter amount To deposit : ")
                        if new_amount.isdigit() :
                            new_amount = int(new_amount)
                            if new_amount > 0 :
                                self.balance += new_amount
                                print("Your amount has been deposited successfully!!!")
                                print(f"your current balance is {self.balance}")
                                self.transaction_history.append(f"{new_amount} deposited ")
                            else:
                                print("Enter Amount higher than 0 !")
                        else:
                            print("Enter amount in Digits")
                    else:
                        print("Incorrect PIN !")
                    
                elif choice == 3 :
                    
                    ver_pin = input("Enter your PIN: ")
                    if ver_pin == self.pin :
                        with_amount = input("Enter amount to withdraw : ")
                        if with_amount.isdigit():
                            with_amount = int(with_amount)
                            if with_amount <= 0 :
                                print("Enter Valid Amount!!")
                            else:
                                if with_amount <= 1000 :
                                    charge = 0
                                elif with_amount <= 20000 :
                                    charge = 100
                                elif with_amount <= 100000 :
                                    charge = 1000
                                elif with_amount > 100000 :
                                    charge = 2000
                                total = with_amount + charge

                                if self.balance >= total :
                                    self.balance = self.balance - total 
                                    print(f"Your current balace is {self.balance}")
                                    self.transaction_history.append(f"{with_amount} withdrawn ({charge} charged) ")
                                else:
                                    print("Insufficient balance including charges")
                        else:
                            print("Enter amount in digits")
                    else:
                        print("Incorrect PIN!!!")
                
                elif choice == 4 :

                    ver_pin = input("Enter PIN :")
                    if ver_pin == self.pin :
                        print(f"Your current balance : {self.balance}")
                    else:
                        print("Incorrect PIN!!")

                elif choice == 5 :

                    ver_pin = input("Enter PIN:")
                    if ver_pin == self.pin :
                        if len(self.transaction_history) == 0 :
                            print("No Transactions to print")
                        else:
                            for index, text in enumerate(self.transaction_history[-10:], start=1):
                                print(index, text)
                
                elif choice == 6:

                    print("Thanks for using ATM machine")
                    break
                    
                else:
                    print("Enter valid choice!!")

atm = ATM()
atm.activate_card()
atm.pin_creation()
atm.initial_deposit()
atm.main_menu()



              
```
