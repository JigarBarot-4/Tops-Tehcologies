## SCENARIO 1
You are deploying a delivery time prediction model for a food delivery startup. The data
science team trained the model in a Jupyter notebook using scikit-learn, and you need
to ship it as part of a production Flask service that will handle several hundred requests
per hour.


## Question: 
Compare Pickle and Joblib as serialisation choices for this deployment context.
Which would you use and why? What risks does each carry when the model is loaded back in an
environment that differs from the one it was trained in?


### Answer:

For this deployment, I would use **Joblib** because it is faster and works better with scikit-learn models, especially when the model contains large NumPy arrays. **Pickle** is a general-purpose serialization library and can also save models, but it may be slower for large files. Both methods can face compatibility problems if the Python version or library versions are different when loading the model. So, it is always better to use the same environment in production as the one used during training.

## SCENARIO 2
You are working on a food delivery app that wants to auto-generate high-quality
promotional images of dishes using only a restaurant's text description and a reference
photo of the plated meal. The marketing team wants realistic, visually rich results.

## Question:
 A colleague suggests using a Variational Autoencoder (VAE), while another
recommends a Stable Diffusion-based model. Compare the two approaches for this
image-generation task — address their key architectural differences, output quality, and which is
more appropriate for a production marketing pipeline.


### Answer:

A **Variational Autoencoder (VAE)** generates images by learning a compressed representation of the data, but the generated images are usually less detailed and slightly blurry. **Stable Diffusion** creates images by gradually removing noise, which produces much more realistic and high-quality results. Since the marketing team wants attractive and realistic food images, Stable Diffusion is the better choice. It can also use both text descriptions and reference images to generate better outputs.

## SCENARIO 3
You are developing an AI assistant for a food delivery platform. The assistant must
handle two distinct jobs: (1) answering open-ended questions such as 'What cuisines
are trending near me this weekend?', and (2) classifying incoming customer reviews as
Positive, Neutral, or Negative for the analytics dashboard.

## Question: 
GPT-style decoder-only models and BERT-style encoder-only models are both
available to your team. Which architecture is better suited to each job, and what fundamental
property of each architecture makes it the right choice for that specific task?


### Answer:

For answering open-ended questions, I would use a **GPT-style decoder model** because it is designed to generate natural and detailed responses. For classifying customer reviews as Positive, Neutral, or Negative, I would use **BERT** because it understands the meaning of the entire sentence before making a prediction. GPT is mainly used for text generation, while BERT is better for text understanding and classification tasks.

## SCENARIO 4
You are building a customer support bot for a food delivery platform. The bot must
handle complaint categories it has never been explicitly trained on — such as 'wrong
item delivered', 'packaging damaged', and 'missing cutlery'. Your team has only 3
annotated complaint–response pairs available per category.
## Question:
 Explain the difference between zero-shot and few-shot prompting. Given the
constraints described, which strategy would you apply, and how would you structure the prompt
to maximise the quality of the bot's responses?


### Answer:

**Zero-shot prompting** means asking the model to perform a task without giving any examples. **Few-shot prompting** means providing a few examples before asking the model to answer. Since only three complaint-response examples are available for each category, I would use few-shot prompting. I would include those examples in the prompt and then ask the model to respond to the new customer complaint in the same style.

## SCENARIO 5
You are designing a LangChain-based food ordering assistant. A simpler alternative is a
single LLM call that takes the user's message and returns a response directly. Your team
is debating whether to build a full agent with tools or use a straightforward LLMChain.

## Question: 
Describe what distinguishes a LangChain agent from a simple LLM chain. Given that
your food ordering assistant must check live restaurant availability, call an API to calculate
estimated delivery time, and validate discount codes, justify which approach is more
appropriate — and identify two risks the agent approach introduces.


### Answer:

A simple **LLMChain** only takes the user's input and generates a response. A **LangChain Agent** can decide which tools to use, such as calling APIs or checking live information before answering. Since the assistant needs to check restaurant availability, calculate delivery time, and validate discount codes, an agent is the better choice. However, agents are more complex to build and may take more time or API calls to complete a task.

## SCENARIO 6
You are optimising a food delivery chatbot that must accurately answer questions
about the platform's 500-page policy document and a menu catalogue that is updated
daily. Your team is evaluating two approaches: fine-tuning the underlying LLM on this
data, or implementing a Retrieval-Augmented Generation (RAG) pipeline.

## Question:
 Compare fine-tuning and RAG for this specific scenario. Address update frequency,
cost, hallucination risk, and data privacy. Which approach is more suitable for a daily-updated,
document-grounded Q&A bot, and what limitations does your recommended approach still
carry?


### Answer:

For this chatbot, I would choose **RAG (Retrieval-Augmented Generation)** instead of fine-tuning. The menu and policy documents change frequently, so RAG can use the latest documents without training the model again. It is also less expensive and reduces the chances of hallucinations because the answers are based on retrieved documents. One limitation is that if the retrieval system does not find the correct information, the model may still give an incomplete or incorrect answer.
