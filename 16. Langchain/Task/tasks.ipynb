## Task 2
Write a Python script that sends a prompt to OpenAI's GPT-3.5 API directly and prints the response. Then, list two problems you faced or could face with this direct API approach.
Hint: Consider error handling, prompt management, or scalability.

```python
from google import genai
from dotenv import load_dotenv
import os

load_dotenv()

api_key = os.getenv("GEMINI_API_KEY2")
print(api_key)  # Remove this after testing

client = genai.Client(api_key=api_key)

prompt = input("Enter your prompt: ")

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)

print(response.text)
```

## Task 3
Use LangChain's PromptTemplate class to create a dynamic prompt that takes a user's favorite IPL team as input and generates a question like 'Why do you support [team] in IPL?'. Print the final prompt for 'Gujarat Titans'

```python
from langchain_core.prompts import PromptTemplate

template = PromptTemplate(
    input_variables=["team"],
    template="Why do you support {team} in IPL?"
)

final_prompt = template.format(team="Gujarat Titans")

print(final_prompt)
```

## Task 4
Build a simple chain in LangChain that takes a user's favorite food (like 'pani puri') and uses an LLMChain to generate a fun fact about it. Print the output.

```python
from dotenv import load_dotenv
import os

from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_classic.chains import LLMChain

load_dotenv()

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    google_api_key=os.getenv("GEMINI_API_KEY2"),
    temperature=0.7
)

prompt = PromptTemplate(
    input_variables=["food"],
    template="Tell me one fun fact about {food}."
)

chain = LLMChain(
    llm=llm,
    prompt=prompt
)

response = chain.invoke({"food": "pani puri"})

print(response["text"])
```
